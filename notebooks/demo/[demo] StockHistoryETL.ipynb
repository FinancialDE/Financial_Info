```python
import os
import sys
sys.path.append('../../')

from financial_etl.stock_history import StockHistory_ETL
```

```python
etl = StockHistory_ETL()
```

```python
symbols = ['JPM', 'GS']
```

## Define Storage Directory

```python
dir_repo = '/Users/hhg/MLAI_Science/projects/finance/repo/HHworking/'
dir_data_lake = dir_repo + 'data_lake/stock_history/'
dir_data = dir_repo + 'data'

if not os.path.exists(dir_data): os.makedirs(dir_data)

if not os.path.exists(dir_data_lake): os.makedirs(dir_data_lake)
```

## Extract

```python
for symbol in symbols:
    filename = os.path.join(dir_data_lake, symbol+'.json')
    etl.extract(symbol, filename)
```

## Transform

```python
filename_out = os.path.join(dir_data, 'stock_history.csv')
print('Transformed file saved to: ', filename_out)

df = etl.transform(symbols, dir_data_lake, filename_out)
df
```
